```python
from scholarly import scholarly
import pandas as pd
import time

# Lista de términos de búsqueda
search_terms = ["bogota care blocks", "manzanas de cuidado bogota"]

# Resultados únicos
results = []
titulos_vistos = set()

# Número de resultados por término
n_resultados = 35

for term in search_terms:
    print(f"\nBuscando: {term}")
    search_query = scholarly.search_pubs(term)

    for i in range(n_resultados):
        try:
            article = next(search_query)
            bib = article.get("bib", {})

            titulo = bib.get("title", "").strip()

            # Evita duplicados
            if titulo in titulos_vistos:
                continue
            titulos_vistos.add(titulo)

            autores = ", ".join(bib.get("author", []))
            anio = bib.get("pub_year", "")
            link = article.get("pub_url", "")
            resumen = bib.get("abstract", "")

            # Filtrar por año
            if anio in ["2024", "2025"] or anio == "":
                results.append({
                    "Año": anio,
                    "Autores": autores,
                    "Título": titulo,
                    "Link": link,
                    "Resumen": resumen
                })

            time.sleep(3)  # Esperar para no ser bloqueado

        except StopIteration:
            break
        except Exception as e:
            print(f"Error: {e}")
            continue

# Convertir a DataFrame
df = pd.DataFrame(results)

# Mostrar resultados
print(df)

# Guardar a Excel
df.to_excel("articulos_bogota_cuidado.xlsx", index=False)
```
